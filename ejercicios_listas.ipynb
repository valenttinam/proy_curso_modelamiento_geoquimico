```python
import matplotlib.pyplot as plt
#EJERCICIO 1: Listas

# 1. Lista con las abreviaturas de los meses (Cadenas de texto / str)
precip_months = ["Ene", "Feb", "Mar", "Abr", "May", "Jun", "Jul", "Ago", "Sep", "Oct", "Nov", "Dic"]

# 2. Lista con los valores de precipitación (Números reales / float)
precip_inches = [0.7, 0.75, 1.85, 2.93, 3.05, 2.02, 1.93, 1.62, 1.84, 1.31, 1.39, 0.84]

#Para verificar usamos print
print("Meses:", precip_months)
print("Precipitación:", precip_inches)
```

```python
#EJERCICIO 2: Convertir pulgadas a milímetros

# 1. Creamos la variable con el factor de conversión
inch_to_mm = 25.4

# 2 Crear una nueva lista con los valores convertidos
#Usando el método .copy() y la pista de indexación (lista[i])
precip_mm = precip_inches.copy()

# Recorrer la lista usando su índice (posición)
for i in range(len(precip_mm)):
    # Multiplico el valor en la posición 'i' por el factor y lo guardamos ahí mismo
    precip_mm[i] = precip_mm[i] * inch_to_mm

# Imprimir ambas listas para comprobar el resultado y que la original no le afecto la conversión
print("Precipitación en pulgadas:", precip_inches)
print("Precipitación en milímetros:", precip_mm)
```

```python
#EJERCICIO 3: Crear una lista de listas

# 1. Crear la lista de listas (agrupando ambas variables)
all_precip_data = [precip_months, precip_mm]

# 2. Imprimir para verificar el resultado
print("Lista de listas completa:")
print(all_precip_data)

# imprimir cuántas sublistas tiene la lista all_precip_data
print("\nNúmero de sublistas adentro:", len(all_precip_data))

# Crear figura y ejes 
fig, ax = plt.subplots(figsize=(8, 6))

# Graficar datos )
ax.bar(
    all_precip_data[0],    # listname_x_axis -> Los meses (índice 0)
    all_precip_data[1],    # listname_y_axis -> Precipitación en mm (índice 1)
    color="blue"           # Cambiamos "red" a "blue"
)

# Configurar títulos y etiquetas
ax.set(
    title="Precipitación Media Mensual",
    xlabel="Meses del año",
    ylabel="Precipitación (mm)"
)

# Mostrar el gráfico
plt.show()
```

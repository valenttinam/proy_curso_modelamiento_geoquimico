```python
import pandas as pd
import matplotlib.pyplot as plt

# PASO 1: Elegir/Crear los Datos (IDEAM - Barranquilla)
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']

# Promedios históricos reales para Barranquilla (Estación Aeropuerto)
temperatura_c = [26.7, 26.8, 27.0, 27.5, 28.1, 28.2, 28.0, 27.9, 27.6, 27.3, 27.2, 26.9]
precipitacion_mm = [1.2, 0.5, 1.6, 24.6, 116.4, 85.3, 67.5, 107.0, 154.2, 164.6, 74.5, 27.4]

df_clima = pd.DataFrame({
    'Mes': meses,
    'Temperatura': temperatura_c,
    'Precipitacion': precipitacion_mm
})
```

```python

# PASO 2: Crear el Gráfico (Orientado a Objetos)
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(10, 8))

# Subgráfico 1: Temperatura vs Tiempo
ax1.plot(df_clima['Mes'], df_clima['Temperatura'], marker='o', color='darkorange', linewidth=2)
ax1.set_title('Promedio Mensual de Temperatura en Barranquilla', fontsize=12, fontweight='bold')
ax1.set_xlabel('Mes del Año', fontsize=11)
ax1.set_ylabel('Temperatura (°C)', fontsize=11)
ax1.grid(True, linestyle='--', alpha=0.6)

# Subgráfico 2: Precipitación vs Tiempo
ax2.bar(df_clima['Mes'], df_clima['Precipitacion'], color='dodgerblue', alpha=0.8)
ax2.set_title('Acumulado Mensual de Precipitación en Barranquilla', fontsize=12, fontweight='bold')
ax2.set_xlabel('Mes del Año', fontsize=11)     
ax2.set_ylabel('Precipitación (mm)', fontsize=11) 
ax2.grid(True, axis='y', linestyle='--', alpha=0.6)

# Título principal y ajuste
fig.suptitle('Análisis Climático Anual: Barranquilla, Colombia (Datos IDEAM)', fontsize=14, fontweight='bold', y=1.02)
fig.tight_layout()

plt.show()
```
